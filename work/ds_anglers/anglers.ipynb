```python
import os
import glob

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
# csvファイル読込
df = pd.DataFrame(columns = [])

for f in glob.glob("data/choka*"):
    tmp = pd.read_csv(f)
    df = pd.concat([df, tmp])
```

```python
# データ編集
df['temperture'] = df['temperture'].str.replace('℃', '')
```

```python
df.head()
```

```python
# 魚種別
df_cnt = df.groupby(by='gyoshu',as_index=False).size().sort_values('size', ascending=False)

fig, ax = plt.subplots(figsize=(40, 20))
plt.rcParams["font.size"] = 30
df_cnt.plot(kind='bar', x='gyoshu', ax=ax)
```

```python
# タイドグラフ別
df_cnt = df.groupby(by=['tide'],as_index=False).size().sort_values('size', ascending=False)

fig, ax = plt.subplots(figsize=(40, 20))
plt.rcParams["font.size"] = 30
df_cnt.plot(kind='bar', x='tide', ax=ax)
```

```python
# 気温別
df_cnt = df.groupby(by=['temperture'],as_index=False).size().sort_values('size', ascending=False)

fig, ax = plt.subplots(figsize=(40, 20))
plt.rcParams["font.size"] = 30
df_cnt.plot(kind='bar', x='temperture', ax=ax)
```
